```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer, MissingIndicator



# simple imputer

# Part of scikit-learn.

# Used to fill missing values in your dataset.

# Works for numerical or categorical data
#  (depending on strategy you choose).



## MissingIndicator

# This adds extra columns (binary) that indicate
# whether a value was missing in the original data.

# Useful because missingness can be informative for the model.

```

```python
v = pd.read_csv('/content/train (2).csv' , usecols= ['Age','Fare','Survived'])
```

```python
v.head()
```

```python
x = v.drop(columns=['Survived'])
y = v['Survived']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
si = SimpleImputer()

#   simple imputer()

# It is a tool from scikit-learn that can replace NaN values in your dataset
# with some calculated value.

# By default:
# It looks for missing values (NaN).
# It fills them with the mean of that column.

                # si

      # si = SimpleImputer()
#  This line creates an object called si
#  which is responsible for filling missing values in your data.


x_train_trf = si.fit_transform(x_train)

# here fit actually means learning the  data and making the formula to replace
# the missing values and tranform are  means to apply the formula to replace
# the missing values ....# so we replace all missing value by the mean value to both train set  and
# testing test too....
# so we replace all missing value by the mean value to both train set  and
# testing test too....

x_test_trf = si.transform(x_test)

# now we take the mean value from train data  set and from that
# mean value we will fill the  mising values too ...
# mean value is stored in si .....its more likely  we study in oops
# now we transform the column of test by the ttrain meean value
# we did not applied  fit in  x test because  we dont have to
#  make my modfel lkearn and derive the faormula or to find out the mean
# value
         # because......

  # If you calculate a new mean or mode from the test data,
  # you cause data leakage (information from the test set
                       #   indirectly gets into model training).
# In the real world, the test data is unseen,
# so you must transform it using only the statistics
# learned from the training data

```

```python
print("Imputation values learned from train:", si.statistics_)
print("Column order:", x_train.columns.tolist())

# this syntax is taken from chatgpt just to
# getting ....to know the mean value .....
```

```python

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train_trf , y_train)

y_pred = clf.predict(x_test_trf)
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

so here what we are doing is to finding out the accuarcy without using missing
indicator ....so we got of accuacy 61 ....now we going to see our accuracy by
planting the feature of missing indicator ....

```python
mi  = MissingIndicator()
# missing indicator is a scikit learn class ...
# herre mi is making a object to generate of column for
# missing value

mi.fit(x_train)
# so basically it learns at which column there is missing value
# or to scan the missing value ...

```

```python
mi.features_

# actually this is the  scikit learn function will return 0 and stored 0
# wherever the missing values this  finds out......
```

```python
x_train_missing = mi.transform(x_train)

# so here we are making  missing indicator ("separate column") for our x train
# data which will give the true false for missing no missing
```

```python
x_train_missing
```

```python
x_test_missing = mi.transform(x_test)
```

```python
x_test_missing

# mi.transform(x_train) returns a NumPy array
#  (or sparse matrix), not a pandas DataFrame.
# This is because scikit-learn transformers by default work
# with NumPy arrays—they are not pandas-aware unless you explicitly wrap them.

# Output shape: (n_samples, n_features_selected)

# Each column represents a feature
# (an indicator for a missing column),
# but it's in array format, not DataFrame.

```

```python
x_train['Age_NA'] = x_train_missing
```

```python
x_train
```

```python
x_test['Age_NA'] = x_test_missing
```

```python
x_test
```

```python
si = SimpleImputer()

x_train_trf2 = si.fit_transform(x_train)
# here x_train_trf2 ::  it is just variable using to store and
# to understand by the
# name that what we are doing here .....

# si.fit
x_test_trf2 = si.transform(x_test)

##                         # reason
#                   why  we did missing indicator
#                    as well as simple imputer too


# If you only impute (fill with mean),
#  the model never knows which values were originally missing.

# Adding MissingIndicator creates extra binary features
# that preserve this information, which can improve accuracy in some cases.

# Then, SimpleImputer fills actual missing values
# so the model can work without NaNs.

# This is a standard practice in machine learning → it gives the model both:
# the filled value (so algorithms don’t break)
#  the missingness signal (which can be important for prediction)

```

```python
x_train_trf2
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train_trf2 , y_train)

y_pred = clf.predict(x_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

```python
                     # shortcut for  making missing indicator
            # as well as for simple imputing

si = SimpleImputer(add_indicator=True)
```

```python
x_train = si.fit_transform(x_train)
x_test = si.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train_trf2 , y_train)

y_pred = clf.predict(x_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
