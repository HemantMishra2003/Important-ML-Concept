```python

# 1. Batch Gradient Descent


#            Batch Gradient Descent

# What is it?
# In Batch Gradient Descent, we use all the training data
# at once to calculate the gradient (slope).

#            How does it work?

#  Imagine you have 1000 data points.
# It uses all 1000 points together to calculate the gradient.
# Then, it updates the weights (slope m and intercept b) just once per round.

# In the next round (epoch), it again uses all 1000 points
# to recalculate and update.

#  Pros:
# Very stable and smooth training
# Moves in the correct direction towards the minimum loss

#  Cons:

# Slow if data is very large
# Needs a lot of memory (since it loads all data at once)


    # How does Gradient Descent work?


# Yes, in Gradient Descent:
# We repeatedly update the slope (m) and intercept (b)
# After every update, the line becomes a little better at fitting the data
# The goal is to ruduce the error (loss) as much as possible
# Finally, we get the best-fit line that gives the lowest error

```

```python
 #  Stochastic Gradient Descent


# SGD (Stochastic Gradient Descent) — Simple Explanation
#    What is it?
# SGD means updating the weights (like m and b) after
# looking at each individual data point — one at a time.



# How does it work?

# Let’s say you have 1000 data points:
# It does not use all 1000 at once (like Batch Gradient Descent).
# Instead, it takes just 1 data point, calculates the gradient (slope).
# Then it updates m and b immediately.
# Then it moves to the next data point, and repeats the same steps.
# It does this for all data points again and again.


# Simple Step-by-Step Flow:


# Take 1 data point
# Calculate the loss (error)
# Find the gradient (slope)
# Update m and b
# Move to the next point
# Repeat the steps again




#                         Advantages (Pros):

# Very fast in practice

# Works well with large datasets

# Updates happen instantly after every data point
# (like real-time learning)

        # Disadvantages (Cons):

# It can be noisy — the line may jump up and down
# Sometimes it's unstable and may not reach the perfect solution smoothly



#  Real-life example for you:

# Imagine you're trying to find a connection
# between students' marks and height.

# You go one-by-one:

# Take 1 student's data

# Improve your line (model)

# Then move to the next student's data
#  That’s exactly how SGD works!

```

```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x , y = load_diabetes(return_X_y=True)
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,
                                                    random_state=2)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test, y_pred)
```

```python
import numpy as np
import random

class MBGDRegressor:

    def __init__(self, batch_size, learning_rate=0.01, epochs=100):
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self, X_train, y_train):




        # Initialize coefficients and intercept

        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            for j in range(int(X_train.shape[0] / self.batch_size)):



                # Random mini-batch sampling

                idx = random.sample(range(X_train.shape[0]), self.batch_size)

                X_batch = X_train[idx]
                y_batch = y_train[idx]



                # Predictions
                y_hat = np.dot(X_batch, self.coef_) + self.intercept_



                # Gradients
                intercept_der = -2 * np.mean(y_batch - y_hat)
                coef_der = -2 * np.dot((y_batch - y_hat), X_batch)



                # Update step
                self.intercept_ -= self.lr * intercept_der
                self.coef_ -= self.lr * coef_der


        print("Final Intercept:", self.intercept_)
        print("Final Coefficients:", self.coef_)


    def predict(self, X_test):
        return np.dot(X_test, self.coef_) + self.intercept_
```

```python
# # Train MBGD

mbr = MBGDRegressor(batch_size=int(x_train.shape[0] / 50),
                    learning_rate=0.01, epochs=100)

mbr.fit(x_train, y_train)

y_pred = mbr.predict(x_test)

print("MBGD R2 Score:", r2_score(y_test, y_pred))
```

```python
# SGD Regressor (Sklearn)

from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor(learning_rate='constant', eta0=0.1)
batch_size = 35

for i in range(100):
    idx = random.sample(range(x_train.shape[0]), batch_size)
    sgd.partial_fit(x_train[idx], y_train[idx])

print("SGD Weights:", sgd.coef_)
print("SGD Intercept:", sgd.intercept_)

y_pred = sgd.predict(x_test)
print("SGD R2 Score:", r2_score(y_test, y_pred))
```
