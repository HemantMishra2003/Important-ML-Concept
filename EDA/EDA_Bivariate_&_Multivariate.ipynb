**Box plot(Numerical - categorical)**

also this is the part of mathmatical statics.... this boxplot library is completely based on the statics(interquartile range ) ....this libary states outliers...or to more make sense it noisy or irrelevent data....its based on the iqr  because it gives median as welll as show the outliers  too  ...why not over the only  mean.. because data could have the noisy data like age  is 20 ,30 ,40 and so on but whhen the age is 100 then its damages the model and predicts the wrong value....

```python
sns.boxplot(data = titanic , x = 'Sex' , y = 'Age' , palette = {'male' :'purple' , 'female':'red'}  )
```

so we could see that outliers are mostly in  at men.......not in women

```python
sns.boxplot(data = titanic , x = 'Sex' , y = 'Age' , hue ='Survived', palette = {1 :'purple' , 0:'red'}  )
```
