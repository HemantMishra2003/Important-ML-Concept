**Euclidean distance**

```python
# covered in notebook ....could revisit too at knn imputer lecture

          # brief notes
#   Euclidean distance is the straight-line distance between two points
#   in multi-dimensional space.

#      How is it used in KNNImputer?

# For each row with missing values,
# the algorithm finds the k closest rows using Euclidean distance.

# Closeness = smaller Euclidean distance.

# Then, it fills the missing value with the average of those neighbors
# (for that feature).

# Scikit-learn uses nan_euclidean_distance,
# which ignores the missing features while calculating distance.


## warning

# If features have different scales
     # (e.g., age in years vs. income in lakhs),
    # Euclidean distance will be biased toward large-scale features.

# Solution:
   # Scale the data first using StandardScaler or MinMaxScaler.

# advantages
# more accurate


# cons
# messy calculation time taken
# while deploying to server all the training set is
# put on the server...
# by this it also comnsumes the memory ....


# conclusion

# for smaller and medieum datframe yes u could go for it ...
# but for large  dataset... yes there is too much calculation have  to do
# also huge nuumber od trainesd dataset have to deploy in server too ....
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
v = pd.read_csv('/content/train (2).csv')[['Age' , 'Pclass' , 'Survived', 'Fare']]
```

```python
v
```

```python
v.isnull().mean()*100
```

```python
x = v.drop(columns = ['Survived'])
y = v['Survived']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 2)
```

```python
x_train.head()
```

```python
knn = KNNImputer(n_neighbors = 3 , weights = 'distance')
# herre nighbors is variable means u could assign to any value
# rto niebors but it will flactuate the accuracy ...
# because ofcourse if assign 10 then it will calculate 10 nieghbors
# and then give the best results ....

 # weights 'distance'

#suppose row 3 any index to row 1 any index have
# the distance of 10 but row 3 to row 4 distance is 50
# ....means row 1 to 3 is more closer than 3 to 4 ....
# here at 1 value is 70 and 4 value is 50 ...
# now what we usually do is to
# just find out the mean like (50 + 70)/ 2 and
# and then impute in vacant index ....

 # but here in "weight distance" what we do is to
 # implemented weighted system...
 #   like (1/10 *70 + 1/50*50)/2




x_train_trf = knn.fit_transform(x_train)

x_test_trf = knn.transform(x_test)

# fit_transform(x_train): Fills missing values in training data
# by finding k-nearest rows within training set.

# transform(x_test): Fills missing values in test data
# using neighbors from training set only (avoids data leakage).



# extra info
# x_train_trf is a NumPy array, not a list of names.
# No string labels for columns or rows.
```

```python
pd.DataFrame(x_train_trf , columns = x_train.columns)

#  DataFrame is a class/function in the pandas library.
#
```

```python
ir = LogisticRegression()

ir.fit(x_train_trf , y_train)

y_pred = ir.predict(x_test_trf)

accuracy_score(y_test , y_pred)*100

#
```
