```python
 # Gradient Descent
          #  an algorithm whom if we give a function then gradient distance
          #  gives us minimima
# or

#"Gradient Descent is an optimization method used to find the
#   lowest (minimum) value of a function by
# taking small steps in the direction of the steepest decrease."


  # function should be diffrentiable at all the points

    # used in linear regression
    # also used in logistic regression


#   Error

#   Error = Difference between actual value and predicted value.

#  Loss
        # is a function of error — it converts that error
        # into a positive scalar value
        # that can be minimized by gradient descent.


 #        gradient descent core concept by example

 # "You are standing at the top of a mountain.
 # You don't have a map in your hands —
 # you can only feel the ground to understand
 #  which direction slopes downward.

# And your goal is to reach the valley below,
# the lowest point, where your model's predictions are most accurate —
# that is, where the loss is minimum."

# Gradient descent is used in machine learning models
# that learn by minimizing error,
# like linear regression, logistic regression, and neural networks.
# It adjusts weights and biases to reduce the loss.
# Models like decision trees and KNN don’t use gradient descent.


 # deep learning
             #  In deep learning  gradient descent is the backbone of
             #  machine learning


# gradient descent is the optimisation technique

```

```python
from sklearn.datasets import make_regression
import numpy as np

# we are calling the make regression function


```

```python
x , y = make_regression(n_samples=4, n_features=1, n_targets = 1,
                        noise=80 , random_state =13)

# here by the use of make regression function we are making
# datasets of 4 datapoints  of one input and one outputs
# here noise 80 they kept just to make data  random ...
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression

# "Its goal is to create the best-fit line with the least error."

# we mean that the Linear Regression model tries to find a line that
# passes through the data in a way
# that the total error (difference between actual and predicted values)
# is as small as possible.

# But the model doesn’t magically give the line immediately.

# we  applied ols here because its in lower dimension
# to know also gets the value of slope and interecepts



```

```python
reg =  LinearRegression()
reg.fit(x,y)

# That means: "Now look at the data
# and learn the best slope (m) and intercept (b) that
# make the line fit the data well."


# we are training my model here without keeping
# reserve test inputs ...


# In linear regression, we want to find a line:
# y = m*x + b
# We don’t know the perfect m and b, so we have to search for them.



#   Gradient Descent Ka Role:

# Gradient Descent is like a game of trial and error —
# we start with some random m and b, and then:
# Check the error (loss)
# See which direction (increase/decrease m or b) makes the loss smaller.
# Take a small step in that direction.
# Repeat steps 1–3 many times until the loss stops decreasing.
```

```python
reg.coef_

# reg.coef_ is an attribute in scikit-learn's
# linear regression model that gives you
# the coefficients (also called weights) learned by the model
# during training.


# reg.coef_ tells you how strongly each feature (input)
# affects the output.

# go to hard notes for more clearity



# why we keep iterating ::

# Because:

# 1st try: random m, b :: high loss
# 10th try: better m, b → smaller loss

# 1000th try: almost perfect m, b →red line

# So yes , we need to iterate many times
# to slowly reach the best-fit line.
```

```python
reg.intercept_

# need to focus over this area
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')

 # this is trying to make the line of best fit line
 # by the use of  m nad b we made a line

 #  goal : we have to make a line above fit line from the
 #         gradient descent
```

```python
# lets apply Gradient descent assuming slope is constant m = 78.35
# and lets assume the starting value for intercept b = 0
# and have to find the value of b by 26..by startinf from the 0

y_pred = ((78.35*x) + 0).reshape(4)

# This is the mathematical equation used in Linear Regression:

# w = 78.35 → is the weight (coefficient) learned during training

# x → is your input feature(s)

# b = 0 → is the intercept (i.e., the value of y when x = 0)

# y_pred → is the predicted output
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red' , label ='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show

# in below  fig :  red line is the ols line
# green line is the gradient line when b value is 0
# now  our goal is to shift green line closer to the red line
# because red line is at 25 ....
```

```python
m = 78.35
b = 0

loss_slope = -2*np.sum(y -m*x.ravel() - b)

# here slope w.r.t. m=−2⋅∑(y−mx−b)
# it is derived in  your  hard notes

loss_slope

# slope we got -209 ...
```

```python

# lets take learning rate =  0.1

lr = 0.1

step_size = loss_slope*lr
# here we know loss slope value ....

# we say step size of product of loss slope and learning rate
```

```python
# calculating the new intercepts means new b

b = b - step_size
b
```

**now we have new b**

```python
# now by doing and iterating  several times we are gonna do this by loop
# and took b = -100 , slope is same  , learning rate is also same


b = -100
m = 78.35
lr = 0.1

epochs = 10  # telling  how  many times we have to iterate

for i in range(epochs):
  loss_slope = -2*np.sum(y -m*x.ravel() - b)
  b = b -(lr*loss_slope)

  y_pred = m*x + b

  plt.plot(x,y_pred)
  plt.scatter(x,y)

# now after 10 iterate we are ta the top of green lines
#  if we take learning rate smaller we could arrive but gradually

# if we take learning rate bigger the there may be problem occurs

```

```python

# we are making the class whom calculate
# the intercept for my given data which should be closer
# to  -0.74

import numpy as np

class GDRegressor:

  def __init__(self, learning_rate, epochs):  # ✅ fix: def_init_ → __init__
    self.m = 55.91
    self.b = -120
    self.lr = learning_rate
    self.epochs = epochs

  def fit(self, x, y):  # ✅ fix: added self
    # calculate the b using GD
    for i in range(self.epochs):
       loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)  # ✅ fix: use self.m and self.b
       self.b = self.b - (self.lr * loss_slope)  # ✅ fix: use self.lr
       print(loss_slope,self.b)
    print(self.b)
```
