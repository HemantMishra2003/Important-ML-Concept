```python
new_v = v[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]

# so here we applied two bracket because we have to bring several columns
# if i have to  bring only one column then we  could aplly
# only one  bracket
```

```python
new_v.dropna(inplace = True)
# removes completely  that  row whom have
# missing values
```

```python
new_v.head(1)
```

```python
#              feature construction


# so by sibsp and parent_child we  have two columns.. so we unite these
# all  the  two colums... and make family_type column

# 0   alone
# 1   small
# 2   large
# by this  sometimes ou model accuarcy gets increased....


#     feature splitting


```

```python
x = new_v.iloc[:,0:4]
# we take x variable to store
y = new_v.iloc[:,-1]
```

```python
x.head(1)
```

```python
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring = 'accuracy',cv = 20)*100)
```

```python
     # Applying Feature Construction
```

```python
x['Family_size']  = x['SibSp'] + x['Parch'] + 1

# creates a new column Family_size.
# It adds siblings/spouse (SibSp) and parents/children (Parch)
# together, plus 1 for the passenger itself.
# This gives the total number of family members
# traveling with that passenger.
```

```python
x.head(1)
```

```python
def myfunc(num):
# def myfunc(num): means we are creating a function
# named myfunc that takes one input called num.
# This function will return 0, 1, or 2
# based on the value of num (alone, small family, or large family).
    if num == 1:
       # alone
        return 0
    elif num>1 and num <= 4:
        return 1    # small family
    else:
        return 2    # large family
```

```python
myfunc(4)
# actually we could see when this defining function
# is returning 1 because we already defined
# that if number is greater than 1 and lesser than 4
# then have to return 1
```

```python
x['Family_type'] = x['Family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns = ['SibSp','Parch','Family_size'],inplace = True)
```

```python
x.head
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring = 'accuracy',cv = 20)*100 )
```
