```python
# f1 score use :


# Class imbalance: e.g., fraud detection, rare disease detection.

# When both False Positives and False Negatives are important.

# When you want a single number that balances
# the trade-off between precision and recall.


# It basically means that if we have Model 1 and Model 2,
# and both have same accuarcy then whom we choose ?

# we will choose the one with the higher F1 score —
# because it shows a better balance between precision and recall.
```

```python
# example model 1

# accuracy  0.9
# precision 0.90
# recall    0.50
# f1        0.64

# vs
# accuracy  0.9
# precision 0.8
# recall    0.8
# f1        0.8

# Model 1: Precision is high, but recall is low → F1 score is moderate.

# Model 2: Precision and recall are balanced
# → F1 score is high → better choice.
```

```python

import numpy as np
import pandas as pd
```

```python

v = pd.read_csv("/content/heart.csv")
```

```python

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(v.iloc[:,0:-1] ,
                                                       v.iloc[:,-1], test_size = 0.2 , random_state= 2)
```

```python


from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python

clf1.fit(x_train , y_train)
clf2.fit(x_train , y_train)
```

```python

y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score , confusion_matrix
print("Accuracy of Logistic regression" , accuracy_score(y_test , y_pred1)*100)
print("Accuracy of Decision Tree" , accuracy_score(y_test , y_pred2)*100)
```

```python
# class imbalance :

#    means some classes have way more samples
#    than others in your dataset..

# Example

# Say you’re building a model to detect if an email is spam or not:

# class          samples
# not spam (0)   9500
# spam  (1)       500

#  Why it’s a problem

# A dumb model that always predicts “Not Spam” will still be 95% accurate,
#  but it’s useless because it never catches spam.

# The model may get biased towards the majority
# class because it sees it more often during training.
```

```python
#    micro pricision vs weighted precision

#  Micro Precision

# What it does:

# Counts total true positives and total false positives
# across all classes first, then applies the precision formula.

# Every prediction is weighted equally,
#  regardless of which class it belongs to.

# Basically: “Overall correctness of positive predictions,
# ignoring class imbalance.”


# When to use:


# When class imbalance doesn’t matter much.

# When you just care about overall model performance
# across all classes combined.

# Good for multi-class classification when you treat every instance equally.
```

```python
# deep insight  if one class prediction is  100 true
# and other class precion is  0 percent false means predict all wrong
# then

# If only one class were important, then looking at per-class precision might not be necessary.
# But in the real world, every class has its own importance — and that’s where the problem comes in.

# Main problem — Model is a master on one side, but fails on the other

# Imagine you build an AI for a doctor that predicts two things:

# 0 = Healthy

# 1 = Cancer

# If your model always predicts “Healthy,” then:

# Precision for Healthy = 100%

# Precision for Cancer = 0%  (because it never detects it)

# The micro average might still be high because most patients are healthy.
# But does that mean the model is good?
# No — because the thing that was more important to detect (Cancer) has 0% performance.

# That’s why per-class precision is important:

# It shows you how well the model performs for each class.

# Otherwise, the performance of the big class will hide the failure
# on the small class.

# In business, healthcare, or fraud detection,
# missing the minority class can be a big risk.

# Simply put:

# It’s not enough to be correct for just one side.
# The model needs to be right for every class,
#  otherwise it’s like being half-blind.
```

```python
# how we make it alright


# Yes, exactly — if Class 1 has 100% precision and Class 2 has 0%,
# that’s a problem,
# because it means the model is perfect for one class
# but completely useless for the other.

# How to fix this — Class imbalance & poor per-class performance

# 1. Balance the data

# Oversampling → Duplicate examples from the minority class (Class 2)
# or use algorithms like SMOTE.

# Undersampling → Remove some examples from the majority class (Class 1)
# so both classes have similar sizes.

# 2. Use class weights

# In training, set class_weight='balanced' (available in scikit-learn).

# This makes the model “punish” mistakes on the minority class more heavily.

# 3. Create better features

# Sometimes the model fails because the features don’t capture
# enough information for Class 2.

# Use feature engineering to make Class 2 patterns more distinct.

# 4. Algorithm tuning

# Adjust the decision threshold so the model detects more of Class 2.

# Try different algorithms (Random Forest, XGBoost, etc.)
# that handle imbalance better.

# Bottom line:

# You have to force the model to pay attention to the minority class (Class 2),
# otherwise it will always stay biased toward the majority class (Class 1).

# If you want, I can make a simple before/after example showing h
# ow adding class weights improves Class 2 precision.
```

```python
# class weight

#            Class weight simply means:

# During training, you tell the model: “Mistakes on the minority
#  class should get a bigger penalty,
# and mistakes on the majority class should get a smaller penalty.”

#  Problem without class weight

# If your data has Class 0 = 90% and Class 1 = 10%,
# the model will focus more on Class 0 because there’s more data for it.
# Result: Class 1 gets ignored.

#  What class weight does

# When you set class_weight='balanced' (in scikit-learn),
# the algorithm assigns weights inversely proportional to class frequency:

# Formula:

# Weight for class
# =
# Total samples
# (Number of classes × Samples in that class)
# Weight for class=
# (Number of classes × Samples in that class)
# Total samples
# 	​


# Example:

# Class 0 (90 samples) → weight ≈ 0.56

# Class 1 (10 samples) → weight ≈ 5.0

# This means one wrong prediction for Class 1
# is punished about the same as nine wrong predictions for Class 0.

#  Benefit

# The model takes both classes more seriously.

# Precision and recall for the minority class can improve significantly.

# If you want, I can show you a small before/after example
# where adding class weights boosts Class 1 precision .
```
