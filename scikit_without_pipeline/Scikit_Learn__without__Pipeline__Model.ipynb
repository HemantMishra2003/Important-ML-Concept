```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
# here we  have droped survived column fromm x axis but
# retained in y axis... because from x axis inputs we have to predict y axis
# and y axis  is  survived column

x_train , x_test , y_train , y_test = train_test_split(v.drop(columns = ['Survived']) , v['Survived'] , test_size = 0.2 , random_state = 42)

```

...by null function we could see that age and emabarked have 177 and 2 missing  values respectivey........

```python
# apllying imputation
si_age = SimpleImputer() # made object for missing values for age
si_embarked = SimpleImputer(strategy = 'most_frequent') # made object for
# embarked for missing values ....strategy syntax means in this... replace missig
# values from the mean in embarked


# in this we are fitting tha data means selecting the mising values then
# transform through the simple imputer class ...imputer trandform
# the missing values
x_train_age = si_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])
# same for x test too
x_test_age = si_age.transform(x_test[['Age']])
x_test_embarked = si_embarked.transform(x_test[['Embarked']])

```

** importing one hot encoder class to transform categorical values or strings values into  numbers **

one hot encoding sex and embaked

```python
ohe_sex = OneHotEncoder(sparse_output = False , handle_unknown = 'ignore')
ohe_embarked = OneHotEncoder(sparse_output = False , handle_unknown = 'ignore')

x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train_embarked)

x_test_sex = ohe_sex.transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.transform(x_test_embarked)
```
