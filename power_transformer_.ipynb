```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer
```

```python
v = pd.read_csv('/content/concrete_data.csv.csv')
```

```python
v.head()
```

lets find out null value have or not

```python
v.isnull().sum()
```

**we are  quite lucky that there is no missing value**

```python
v.describe()
```

threre is no minus value but there is yes min value 0 have ...by

```python
x = v.drop(columns = ['Strength'])
y = v.iloc[: , -1]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

**we are goona apply regression algorithm.. without any transformation  neither we applied**
**even scaling**

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
y_pred = lr.predict(x_test)
r2_score(y_test , y_pred)*100
## in linear regression accuracy comes out by the r2_score
## while in logical regression accuracy comes out by the accuracy_score
```

but how r2 square works it still a big question !!


**lets check by  cross value score**

```python
lr = LinearRegression()
np.mean(cross_val_score(lr , x , y , scoring = 'r2')*100)
```

since in upper syntax we have not defined cv number then by default it assumes we have defined 5 times....

**goona plot the distplot without any transformation**

```python
for col in x_train.columns:
  plt.figure(figsize = (14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])
  plt.title(col)

  plt.subplot(122)
  stats.probplot(x_train[col] , dist = "norm" , plot = plt)
  plt.show()

  plt.title(col)
```

**now we are goona do box plot to bring the messed data in normal distribution
although we should have been done the scaling to get more accuarate result ...
but ;lets leave that for now ...**

```python
# applying the box clot
pt = PowerTransformer(method = 'box-cox')\
#here we have applied 0.0001
x_train_transformed = pt.fit_transform(x_train + 0.00001)
x_test_transformed = pt.transform(x_test+0.00001)

pd.DataFrame( {'cols':x_train.columns,'box_cox_lambdas':pt.lambdas_})
# here pt. lamdas is  showing the value lamdas for every column
# for more info go to copy  statics section in power transformer column
```

```python
# applying linear regression on transformed data
lr = LinearRegression()
lr.fit(x_train_transformed , y_train)
y_pred = lr.predict(x_test_transformed)
r2_score(y_test , y_pred)*100
# we could see that after applying the
# transformation howe much we have increased the accuracy
# trnsformation by box - cox -------
```

```python
# using cross val score...

pt = PowerTransformer(method = 'box-cox')
x_train_transformed = pt.fit_transform(x_train + 0.00001)

lr = LinearRegression()
np.mean(cross_val_score(lr , x_train_transformed , y_train , scoring = 'r2')*100)
```

**Before transformation vs after transformation  plotting**


**transformation in box plot**

```python
# before and after comparision for box plot
x_train_transformed = pd.DataFrame(x_train_transformed,columns = x_train.columns)
# in above syntax x transform it is numpy array ,
# so we are changing this in  data_frame by which we could
# graph plot

for col in x_train_transformed.columns:
  plt.figure(figsize = (14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])
  plt.title(col)

  plt.subplot(122)
  sns.distplot(x_train_transformed[col])
  plt.title(col)

  plt.show()
```
